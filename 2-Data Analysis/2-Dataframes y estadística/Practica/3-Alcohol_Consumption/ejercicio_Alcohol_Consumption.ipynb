Sabéis que nos preocupa mucho la salud de los ciudadanos. Por eso nos han encargado hacer un análisis sobre el consumo de alcohol.

Nos interesa conocer el consumo de alcohol

1.- por tipo de bebida alcohólica para los distintos continentes y países. 

2.- Además, sabiendo que el vino y la cerveza son los sospechosos habituales, podéis disponer estadísticas sobre estos dos casos en particular.

    1.- Obtener el total de consumo por contiente por categoria
    2.- Obtener los 5 paises con mas consumo por categoria

    4.- Mostrar graficos








```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv"

# Leer el archivo CSV desde la URL y convertirlo en un DataFrame
datos = pd.read_csv(url, sep=",")

# Mostrar las primeras filas del DataFrame
datos.head(25)
```

```python
datos.columns
```

```python
datos.info()
```

```python
# Reemplazar los valores NaN en la columna 'continent' con 'Sin Continente'
datos['continent'] = datos['continent'].fillna('Sin Continente')
```

```python
datos.info()
```

```python
df_consumo_alcohol = pd.DataFrame(datos)
```

```python
df_consumo_alcohol.describe().T 
```

```python
df_consumo_alcohol.groupby(['continent', 'beer_servings'])['beer_servings'].describe().reset_index()
```

```python
continent_counts = df_consumo_alcohol.groupby('continent').agg(
    beer_servings_sum=('beer_servings', 'sum'),
    beer_servings_mean=('beer_servings', 'mean'),
    spirit_servings_sum=('spirit_servings', 'sum'),
    spirit_servings_mean=('spirit_servings', 'mean'),
    wine_servings_sum=('wine_servings', 'sum'),
    wine_servings_mean=('wine_servings', 'mean'),
    total_litres_of_pure_alcohol_sum=('total_litres_of_pure_alcohol', 'sum'),
    total_litres_of_pure_alcohol_mean=('total_litres_of_pure_alcohol', 'mean')
).reset_index()
continent_counts
```
