# Consumo de alcohol

Vamos a intentar sacar ciertas ideas de nuestro consump de acohol por país y continente.

```python
import numpy as np
import pandas as pd

data = pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv")
data
```

```python
data.describe(include="all")
```

Parece que tenemos datos faltantes en continente.

```python
data["total_bws"] = data["beer_servings"] + data["wine_servings"] + data["spirit_servings"]
data
```

```python
data.loc[data["continent"] == "AF", "total_bws"] = -1
data
```

```python
data["continent"].unique()
```

```python
data.loc[data["continent"].isna()]
```

Parece que todos son del norte de America.

```python
data.loc[data["continent"].isna(), "continent"] = 'AM'
```

```python
data["continent"].unique()
```

Ya tenemos toda la información completa.

```python
data.groupby(["continent", "country"]).mean()
```

```python
mean_consumption = data.groupby(["continent", "country"]).mean()
mean_consumption = mean_consumption.reset_index()
mean_consumption
```

```python
mean_consumption.pivot_table(["beer_servings","wine_servings"], index="country", columns="continent").fillna(0).T
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.bar(mean_consumption["continent"], mean_consumption["beer_servings"])
```

```python
fig, ax = plt.subplots()
ax.bar(mean_consumption["continent"], mean_consumption["beer_servings"], label="Cerveza")
ax.bar(mean_consumption["continent"], mean_consumption["wine_servings"], label="Wine")
plt.legend()
```

```python
agregado = mean_consumption.groupby('continent')[['beer_servings',"wine_servings"]].sum()
agregado
```

```python
import matplotlib.pyplot as plt
import numpy as np

width = 0.50  # the width of the bars
multiplier = 0
continentes = agregado.index
x = np.arange(len(continentes))

fig, ax = plt.subplots(layout='constrained')

for column in agregado.columns:
    offset = width * multiplier
    rects = ax.bar(x+offset, agregado[column], width, label=column.split("_")[0])
    ax.bar_label(rects, padding=2)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Consumption (ltr)')
ax.set_title('Continent')
ax.set_xticks(x + width, continentes)
ax.legend(loc='upper left', ncols=2)

plt.show()
```

```python
eu_consumption = mean_consumption[mean_consumption["continent"] == "EU"]

fig, ax = plt.subplots(figsize=(20,10))
ax.bar(eu_consumption["country"], eu_consumption["beer_servings"])
ax.tick_params(axis='x', rotation=45)
ax.set_ylabel('Consumption (ltr)')
ax.set_title('Continent')

plt.show()
```
