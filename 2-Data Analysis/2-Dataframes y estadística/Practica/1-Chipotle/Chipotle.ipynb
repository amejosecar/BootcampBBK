### Step 2. Tomaremos el fichero en [esta URL](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv) y lo leeremos como un dataframe.

Pista: [read_csv](https://numpy.org/doc/stable/user/absolute_beginners.html#importing-and-exporting-a-csv)

```python
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"

# Leer el archivo CSV desde la URL y convertirlo en un DataFrame
df_chipotle = pd.read_csv(url, sep="\t")

# Mostrar las primeras filas del DataFrame
df_chipotle.head(25)
```

```python
#1ero voy a limpiar la columna item_price quitandole el signo $
df_chipotle_clean = df_chipotle.copy() 

df_chipotle_clean['item_price'] = df_chipotle_clean['item_price'].str.replace('$', '').astype(float)

df_chipotle_clean.head(25)
```

```python
df_chipotle_clean.describe()
```

### Step 3. Veamos los tipos de datos. ¿Podríamos obtener el producto de mayor precio?

```python
# Calcula el precio unitario
df_chipotle_clean['unit_price'] = df_chipotle_clean['item_price'] / df_chipotle_clean['quantity']

# Encuentra el item_name con el precio unitario más alto
max_unit_price_row = df_chipotle_clean.loc[df_chipotle_clean['unit_price'].idxmax()]

# Extrae el item_name
max_unit_name_item = max_unit_price_row['item_name']
max_unit_price_item = max_unit_price_row['item_price']

print(f"El item_name con el item_price más alto por unidad es: {max_unit_name_item} {max_unit_price_item}")
```

### Step 4. ¿Qué productos cuestan más de $10?

```python
### Step 4. ¿Qué productos cuestan más de $10?

# Filtra los registros con unit_price mayor a 10
df_filtered = df_chipotle_clean[df_chipotle_clean['unit_price'] > 10]

# Agrupa por 'item_name' y 'unit_price' y cuenta las ocurrencias
grouped_df = df_filtered.groupby(['item_name', 'unit_price']).count()

grouped_df
```
