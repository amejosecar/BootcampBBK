### Step 2. Tomaremos el fichero en [esta URL](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv) y lo leeremos como un dataframe.

Pista: [read_csv](https://numpy.org/doc/stable/user/absolute_beginners.html#importing-and-exporting-a-csv)

```python
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"

# Leer el archivo CSV desde la URL y convertirlo en un DataFrame
df_chipotle = pd.read_csv(url, sep="\t")

# Mostrar las primeras filas del DataFrame
df_chipotle.head(25)
```

```python
#1ero voy a limpiar la columna item_price quitandole el signo $
df_chipotle_clean = df_chipotle.copy() 

df_chipotle_clean['item_price'] = df_chipotle_clean['item_price'].str.replace('$', '').astype(float)

df_chipotle_clean.head(25)
```

```python
### Step 3. Veamos los tipos de datos. ¿Podríamos obtener el producto de mayor precio?

#aqui obtengo la fila a traves del indice
indice_max_precio = df_chipotle_clean['item_price'].idxmax()

indice_max_precio

producto_mas_caro = df_chipotle_clean.loc[indice_max_precio]

producto_mas_caro
```

### Step 4. ¿Qué productos cuestan más de $10?

```python
### Step 4. ¿Qué productos cuestan más de $10?

# aqui voy a filtrar las filas donde quantity es igual a 1 y item_price es mayor o igual a $10
productos_filtrados_cantidad_precio = df_chipotle_clean[(df_chipotle_clean['quantity'] == 1) & (df_chipotle_clean['item_price'] >= 10)]

# Calcular el total de esos productos
total_productos = len(productos_filtrados_cantidad_precio)

print(f"Son {total_productos} productos con precios mayores de $10")
```

```python
### Step 4.3: ¿Y en cuántos pedidos se ha pagado más de 10$ por un mismo producto? ¿Es lo mismo?

#aqui creo el grupo por item_name y que posean el mismo item_price
productos_agrupados = df_chipotle_clean.groupby(['item_name', 'item_price']).size().reset_index(name='count')

#aqui filtro a los item_price mayor a 10
productos_agrupados = productos_agrupados[productos_agrupados['item_price'] > 10]

productos_agrupados
```

### Step 5. ¿Qué precio tiene cada producto en distintos pedidos? ¿Hay productos con varios precios?

```python
### Step 5. ¿Qué precio tiene cada producto en distintos pedidos? 

#aqui tengo que desglosar los quantity que son mayores a 1, ya que debo contar estos productos tambien 
# por separados para saber si poseen precios distintos.
#entonces pense en un apoyo y busque la forma de iterar dentro de un dataframe.

#esto fue lo que encontre


# Crear un diccionario para almacenar los precios únicos por producto
precios_unicos_por_producto = {}

# Iterar sobre cada fila del DataFrame
for index, row in df_chipotle_clean.iterrows():
    # Obtener el nombre del producto y el precio
    producto = row['item_name']
    precio = row['item_price']
    # Calcular el precio por unidad si la cantidad es mayor a 1
    if row['quantity'] > 1:
        precio /= row['quantity']
    # Agregar el precio a la lista correspondiente en el diccionario
    if producto in precios_unicos_por_producto:
        if precio not in precios_unicos_por_producto[producto]:
            precios_unicos_por_producto[producto].append(precio)
    else:
        precios_unicos_por_producto[producto] = [precio]

# Crear una lista de tuplas para almacenar los datos
datos = []

# Iterar sobre el diccionario y agregar los datos a la lista de tuplas
for producto, precios in precios_unicos_por_producto.items():
    if len(precios) > 1:
        for precio in precios:
            datos.append((producto, precio))

# Crear el DataFrame a partir de la lista de tuplas
df_resultado = pd.DataFrame(datos, columns=['item_name', 'item_price'])

# Mostrar el DataFrame resultante
df_resultado.head(25)
```

```python
### Step 5.b. ¿Hay productos con varios precios?

# Filtro el DataFrame a los productos con múltiples precios
productos_varios_precios_df = df_chipotle_clean[df_chipotle_clean.duplicated(subset=['item_name', 'item_price'], keep=False)]

# Mostrar el DataFrame resultante
productos_varios_precios_df
```

```python
#aqui cuento el total de productos con varios precios
total_contado_varios = len(productos_varios_precios_df)

print(f"El total son {total_contado_varios} productos con varios precios")
```

### Step 7. ¿Cuantas veces se ha pedido los productos más caros?

```python
### Step 7. ¿Cuantas veces se ha pedido los productos más caros?
# 
# aqui busco el precio mas caro
precio_maximo = df_chipotle_clean['item_price'].max()

# aqui cuento cuántas veces aparece el precio máximo en la columna 'item_price'
veces_producto_mas_caro = (df_chipotle_clean['item_price'] == precio_maximo).sum()

print("Cantidad de veces que se ha pedido el producto más caro:", veces_producto_mas_caro)



```
