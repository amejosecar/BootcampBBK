### Step 2. Tomaremos el fichero en [esta URL](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv) y lo leeremos como un dataframe.

```python
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"

data = pd.read_csv(url, sep="\t")
```

```python
data.dropna().describe(include="all")
```

### Step 4. ¿Qué productos cuestan más de $10?

```python
data["item_price"] = data["item_price"].str.replace("$","").astype(float)
```

```python
data.describe()
```

```python
condicion = (data["item_price"] > 10)
data[condicion]
```

```python
data["precio_unidad"] = data["item_price"] / data["quantity"]
condicion = (data["precio_unidad"] > 10)
data[condicion]
```

### Step 4.1: ¿Y cuántos pedidos se han hecho con un producto de más de 10$? ¿Es lo mismo?

```python
precio_maximo = data["precio_unidad"].max()
data[data["precio_unidad"] == precio_maximo]
```

```python
condicion = (data["precio_unidad"] > 10)
data[condicion]["order_id"].nunique()
```

### Step 4.2: ¿Y cuántos pedidos se han hecho de más de 10$? ¿Es lo mismo?

```python
data[data["item_price"] > 10]["order_id"].nunique()
```

### Step 4.3: ¿Y en cuántos pedidos se ha pagado más de 10$ por un mismo producto? ¿Es lo mismo?

```python
productos_distintos = data["item_name"].unique()
for producto in productos_distintos:
    tantos = data[(data["item_name"] == producto) & (data["precio_unidad"] > 10)]["order_id"].nunique()
    if tantos != 0:
        print(f"{producto} se ha pagado a más de $10 en {tantos} pedidos")
```

```python
data_agregada = data.groupby("item_name").aggregate({"item_price": 'mean', 'precio_unidad': 'mean', 'quantity' : 'max', 'order_id': 'count'})
```

```python
data_agregada[data_agregada["precio_unidad"] > 10]
```

### Step 5. ¿Qué precio tiene cada producto en distintos pedidos? ¿Hay productos con varios precios?

```python
data_agregada.groupby(["item_name", "order_id"])[["item_price", "precio_unidad"]].mean()
```

```python
data.pivot_table("item_price", index="item_name", columns="order_id").fillna(0).std("columns")
```
